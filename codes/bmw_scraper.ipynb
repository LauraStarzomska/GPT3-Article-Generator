```python
soup = BeautifulSoup(doc, "html.parser")


#title of article
title_article = soup.select('h3 a')
titles = []
for title in title_article:
  titles.append(title.get_text())

#titles = list(map(lambda x: x.replace("\xa0", ""), titles))
titles = [t for t in titles if len(t) > 5]


#content of article
article_content = soup.select('.serif')
content = []
for x in article_content:
  content.append(x.get_text())


content = list(map(lambda x: x.replace("+++", ""), content))
```

```python
print(len(titles), len(content))
```

```python
full_df = pd.DataFrame(columns=['title', 'content'])

df_current = pd.DataFrame({
            'title': titles,
            'content': content,
            })
        
full_df = full_df.append(df_current, ignore_index = True)
```

```python
#Now we have to go into every article to get the full content of the article

#links = soup.find_all("div", class_ = "text")
links = soup.select('h3')
url = []
for t in links:
  for item in t.find_all('a', href = True):
    url.append(item["href"])
    #print(item)
for i in range(len(url)):
    url[i] = "https://www.press.bmwgroup.com/" + url[i]
```

```python
url
```

```python
#print(url)
articles_intro = []
for link in url:
  r = requests.get(link)
  soup = BeautifulSoup(r.content, "html.parser")
  #message = soup.select("#article-text.serif")
  #message = soup.find_all("article-text", class_ = "serif")
  #message = soup.select("article-text p")
  #message = soup.select("#article-text p , .clear")
  #message = soup.select("p")
  intros = soup.find_all("h2", class_ = "teaser clear")
  for m in intros:
    articles_intro.append(m.get_text())
```

```python
articles_intro
```

```python
print(url)
content = []
for link in url:
  r = requests.get(link)
  soup = BeautifulSoup(r.content, "html.parser")
  text = soup.select("#article-text p , .clear")
  for m in text:
    content.append(m)
```

```python
content
```

```python
df = pd.DataFrame(columns=['header', 'text'])

for i in range(len(content)):
    if content[i].name == 'h2':
        header = content[i].text
        text = ''
        for j in range(i+1, len(content)):
            if content[j].name == 'h2':
                break
            else:
                text += content[j].text
        df = df.append({'header': header, 'text': text}, ignore_index=True)

df
```
